```python
import pandas as pd
import numpy as np
```

```python
a=pd.read_csv("Churn_Modelling.csv")
```

```python
df=pd.DataFrame(a)
```

```python
df.head()
```

```python
a=pd.get_dummies(df['Geography'],drop_first=True)
b=pd.get_dummies(df['Gender'],drop_first=True)
```

```python
df1=pd.concat([df,a,b],axis='columns')
```

```python
df1.head()
```

```python
df1.drop(['RowNumber','CustomerId','Surname','Geography','Gender'],axis='columns',inplace=True)
```

```python
df1.head()
```

```python
X=df1.drop(['Exited'],axis='columns')
y=df1['Exited']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit_transform(X_train)
scaler.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
cr=classification_report(y_pred,y_test)
print(cr)
cm=confusion_matrix(y_pred,y_test)
print(cm)
ac=accuracy_score(y_pred,y_test)
print(ac)
```

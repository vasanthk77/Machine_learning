```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets  import make_classification
```

```python
from sklearn.model_selection import train_test_split
X,y=make_classification(n_samples=2000,n_features=20,n_classes=2,weights=[1,1],random_state=1)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
#### RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier()
model1.fit(X_train, y_train)
y_pred=model1.predict_proba(X_train)
print('rando forest roc-auc score{}'.format(roc_auc_score(y_train,y_pred[:,1])))  ## y_pred we got 1 and 0 values we take 0 values and done that
y_pred_test=model1.predict_proba(X_test)
print('rando forest roc-auc score{}'.format(roc_auc_score(y_test,y_pred_test[:,1])))
```

```python
## LogisticRegression
from sklearn.linear_model import LogisticRegression
model2= LogisticRegression()
model2.fit(X_train, y_train)
y_pred=model2.predict_proba(X_train)
print('Logistic Regression roc-auc score{}'.format(roc_auc_score(y_train,y_pred[:,1])))  ## y_pred we got 1 and 0 values we take 0 values and done that
y_pred_test=model2.predict_proba(X_test)
print('Logistic Regression roc-auc score{}'.format(roc_auc_score(y_test,y_pred_test[:,1])))
```

```python
## Adaboost Classifier
from sklearn.ensemble import AdaBoostClassifier
model3=AdaBoostClassifier()
model3.fit(X_train, y_train)
y_pred=model3.predict_proba(X_train)
print('ada boost roc-auc score{}'.format(roc_auc_score(y_train,y_pred[:,1])))  ## y_pred we got 1 and 0 values we take 0 values and done that
y_pred_test=model3.predict_proba(X_test)
print('ada boost roc-auc score{}'.format(roc_auc_score(y_test,y_pred_test[:,1])))
```

```python
### KNN
from sklearn.neighbors import KNeighborsClassifier
model4=KNeighborsClassifier()
model4.fit(X_train, y_train)
y_pred=model4.predict_proba(X_train)
print('KNN roc-auc score{}'.format(roc_auc_score(y_train,y_pred[:,1])))  ## y_pred we got 1 and 0 values we take 0 values and done that
y_pred_test=model4.predict_proba(X_test)
print('KNN roc-auc score{}'.format(roc_auc_score(y_test,y_pred_test[:,1])))
```

```python
### Noow Here comes
pred=[]
for i in [model1,model2,model3,model4]:
  pred.append(pd.Series(i.predict(X_test)))
```

```python
pred = []
for model in [model1, model2, model3, model4]:
        probabilities = model.predict_proba(X_test)
        pred.append(pd.Series(probabilities[:, 1]))
final_predictions=pd.concat(pred,axis=1).mean(axis=1)
print('ensemble test roc-auc {}'.format(roc_auc_score(y_test,final_predictions)))  # Indexing the probability of class 1
```

```python
fpr,tpr,thresholds=roc_curve(y_test,final_predictions)
thresholds ## calcauting the threshold values
```

```python
from sklearn.metrics import accuracy_score
accuracy_is=[]
for thre in thresholds:
  y_pred1=np.where(final_predictions>thre,1,0)
  accuracy_is.append(accuracy_score(y_test,y_pred1,normalize=True))
```

```python
accuracy_is=pd.concat([pd.Series(thresholds),pd.Series(accuracy_is)],axis=1)
```

```python
accuracy_is.columns=['thresholds','accuracy']
accuracy_is.sort_values(by='accuracy',ascending=False,inplace=True)
accuracy_is.head()
```

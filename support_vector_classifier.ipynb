```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=pd.read_csv("Churn_Modelling.csv")
```

```python
df=pd.DataFrame(a)
```

```python
df.head()
```

```python
df.drop(['RowNumber','CustomerId','Surname'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df1=pd.get_dummies(df['Gender'])
```

```python
df2=pd.get_dummies(df['Geography'])
```

```python
df3=pd.concat([df,df1,df2],axis='columns')
```

```python
df3.head()
```

```python
df3.drop(['Geography','Gender','Spain'],axis='columns',inplace=True)
```

```python

X=df3.drop(['Exited'],axis='columns')
y=df3['Exited']
```

```python
## train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
## using standarization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_scaled=sc.fit_transform(X_train)
X_test_scaled=sc.transform(X_test)
```

```python
## Model creation
from sklearn.svm import SVC
model=SVC(kernel='linear')
model.fit(X_train_scaled,y_train)
```

```python
y_pred=model.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
cr=classification_report(y_test,y_pred)
print(cr)
score=accuracy_score(y_test,y_pred)
print(score)
```
